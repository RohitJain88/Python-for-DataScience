```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x =np.linspace(0,10,10)
y = x**2
```

```python
#Functional
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig = plt.figure()

axes= fig.add_axes([0.1,0.1,0.8,0.8]) # axes parameter - left,bottom, width, height in % in range from 0 to 1
axes.plot(x,y)
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
figs,axes = plt.subplots(nrows=1,ncols=2) #Here figs,axes is tuple unpacking in general we r combining the above two line of code for fig and axes

plt.tight_layout() #Avoid Overlapping

#for curr_ax in axes:
 #   curr_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('Left Plot')
axes[1].plot(y,x)
axes[1].set_title('Right Plot')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=10)#center position
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=2,alpha=0.5,ls='steps') #RGB Hex value, linewidth as lw, linestyle as ls
```
